```python
##“Comparative Trend Analysis of Bitcoin Price Components Over Time”(Line Plot)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))

price_cols = ['Open', 'High', 'Low', 'Close']

for i, col in enumerate(price_cols, 1):
    plt.subplot(2, 2, i)
    plt.plot(df['Date'], df[col], linewidth=2)
    
    plt.title(f'{col} Price Trend', fontsize=12)
    plt.xlabel('Date')
    plt.ylabel('Price')
    
    plt.grid(alpha=0.3)

plt.suptitle('Comparative Analysis of Bitcoin Price Components', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
###Insight
##Bitcoin shows a long-term upward trend with multiple sharp spikes and corrections.
##This indicates speculative growth behavior with high market sensitivity.
```

```python
##“Impact of Log Transformation on Closing Price: Revealing Hidden Trends Beyond Exponential Growth"
```

```python
plt.figure(figsize=(20,8))


plt.subplot(1,2,1)
data['Close'].plot()
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.title('No scaling')
plt.grid(alpha=0.3)

plt.subplot(1,2,2)
np.log1p(data['Close']).plot()
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.grid(alpha=0.3)


plt.title('Log scaling')
plt.yscale('log')
```

```python
##No scale → shows actual price magnitude (good for value understanding)
##Log scale → shows true growth behavior and trend clarity
##This combination helps distinguish between perceived spikes vs actual growth patterns—a key insight for both analysis and decision-making.
```

```python
###“Multi-Timeframe Closing Price Analysis: Identifying Macro Trends, Seasonal Patterns, and Short-Term Fluctuations”
```

```python
plt.figure(figsize=(12,6))
data['Close'].resample('YE').mean().plot()
plt.title('Yearly Average Closing Price Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Price', fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,6))
data['Close'].resample('QE').mean().plot()
plt.title('Quaterly Average Closing Price Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Price', fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,6))

data['Close'].resample('ME').mean().plot()

plt.title('Monthly Average Closing Price Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Price', fontsize=12)

plt.grid(alpha=0.3)
plt.show()
```

```python
###Yearly → Direction (Growth vs Decline)
##Quarterly → Momentum shifts
##Monthly → Volatility & entry/exit signals

##Combining all three gives a complete picture of market behavior across time horizons, crucial for both analysis and decision-making.
```

```python
####“Daily Change Analysis of Closing Price: Measuring Short-Term Volatility and Market Reactivity”
```

```python
data['Close'].pct_change()*100
```

```python
data['Close_price_pct_change']=data['Close'].pct_change()*100
```

```python
data['Close_price_pct_change']
```

```python
plt.figure(figsize=(12,6))
data['Close_price_pct_change'].plot()
plt.title('Daily Percentage Change in Closing Price')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.grid(alpha=0.3)
plt.show()
```

```python
##Insights
##Helps identify risk level in day-to-day trading
##Useful for detecting:
##Volatility intensity
##Market instability periods
##Potential trading opportunities
```

```python
##“Percentage Change Analysis of Closing Price: Tracking Relative Volatility and Market Momentum Over Time”
```

```python
import plotly.express as px

fig = px.line(
    data,
    x=data.index,
    y='Close_price_pct_change',
    title='Daily % Change in Bitcoin Closing Price',
    labels={
        'x': 'Date',
        'Close_price_pct_change': 'Percentage Change (%)'
    }
)

fig.update_layout(
    template='plotly_white',
    title_x=0.5,  # center title
    title_font=dict(size=20),
    xaxis_title='Date',
    yaxis_title='% Change',
    hovermode='x unified'
)

fig.update_traces(
    line=dict(width=2),
    hovertemplate='Date: %{x}<br>% Change: %{y:.2f}%'
)

fig.show()
```

```python
##Insights
##“Percentage change analysis highlights true market volatility and momentum,
##revealing frequent short-term fluctuations with occasional extreme sentiment-driven movements.”
```
